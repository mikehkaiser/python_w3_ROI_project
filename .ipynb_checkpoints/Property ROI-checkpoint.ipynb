```python
# take in purchase price, closing costs, and cost for improvements = initial investment
class ROI:
    def __init__(self, initial_investment = None, income = None, expenses = None):
        self.initial_investment = initial_investment
        self.income = income
        self.expenses = expenses

# ask for investment numbers:
    def cashInvestment(self):
        upfront_costs = []
        home_price = int(input('What was the contract price of the property?'))
        cash_pay = input('Did you purchase the home for cash?')
        if cash_pay.lower() == 'yes':
            upfront_costs.append(home_price)
        elif cash_pay.lower() != 'yes' and cash_pay.lower() != 'no':
            print("Please enter 'Yes' or 'No'")
        else:
            down_pay = int(input('What was your down payment amount?'))
            upfront_costs.append(down_pay)
        closing_costs = int(input('How much were closing costs?'))
        upfront_costs.append(closing_costs)
        initial_repairs = int(input('What was your repair budget?'))
        upfront_costs.append(initial_repairs)
        self.initial_investment = sum(upfront_costs)
        return f'Initial Investment: ${round(self.initial_investment,2)}.'
    
    def monthlyExpenses(self):
        expenses = []
        prop_tax = int(input('How much would property taxes be for the year?'))/12
        expenses.append(prop_tax)
        prop_insurance = int(input('How much would property insurance be for the year?'))/12
        expenses.append(prop_insurance)
        hoa = input('Is the property part of an HOA?')
        if hoa.lower() == 'yes':
            hoa_fees = int(input('What is the monthly HOA fee for the property?'))
            expenses.append(hoa_fees)
        elif hoa.lower() != 'yes' and hoa.lower() != 'no':
            print("Please enter 'Yes' or 'No'")
        landscaping = input('Will you pay for lawn care? (Yes/No)')
        if landscaping.lower() == 'yes':
           land_fee = int(input('How much will landscaping cost each month?'))
           expenses.append(land_fee)
        if landscaping.lower() != 'yes' and landscaping.lower() != 'no':
            print("Please enter 'Yes' or 'No'")
        building_maint = int(input('How much will you pay for building maintenance each month?'))
        expenses.append(building_maint)
        vacancy = input('Are you setting aside money in case of a vacancy? 5% is recommended. (Yes/No)')
        if vacancy.lower() == 'yes':
            vacancy_amt = int(input('How much are you setting aside in case of vacancy? (Enter dollar amount)'))
            expenses.append(vacancy_amt)
        misc_expenses = int(input("Enter any additional expenses or '0' if none."))
        self.expenses = sum(expenses)
        return f'Monthly Expenses are ${round(self.expenses,2)}.'
    
    def monthlyIncome(self):
        mo_income = []
        rental_income = int(input('How much will you receive in rent each month?'))
        mo_income.append(rental_income)
        storage = input('Will the tenant pay for storage on the property? (Yes/No)')
        if storage.lower() == 'yes':
            storage_cost = int(input('What will the tenant pay for storage?'))
            mo_income.append(storage_cost)
        elif storage.lower() == 'no':
            pass
        elif storage.lower() != 'yes' and storage.lower() != 'no':
            print("Please enter 'Yes' or 'No'")
        extra_income = int(input('Please enter any additional income. Enter dollar amount.'))
        mo_income.append(extra_income)
        self.income = sum(mo_income)
        return f'The monthly income for the property is ${round(self.income,2)}.'
    
    def calculateROI(self):
        #capture Investment info
        self.cashInvestment()
        self.monthlyExpenses()
        self.monthlyIncome()
        roi = ((self.income - self.expenses)/self.initial_investment)*100
        return f'Your cash-on-cash ROI is {round(roi,1)}%'
                
house = ROI()
house.calculateROI()                     
                       
                       
# ask for monthly and yearly expenses:
# mortgage payment
# maintenance
# property tax
# homeowners insurance cost
# HOA fees
# Building maintenance
# Vacancy savings
# Repairs
# Capital Expenditures
# Management fees
# # add it together as total monthly expenses

# # ask for income info:
# Rent
# Storage
# Pet rent
# add it together as total monthly income


#cash flow = income - expenses
# subtract, then multiply by 12 to get yearly cash flow/profit

# divide yearly expense by total investment and multiply x 100
```
