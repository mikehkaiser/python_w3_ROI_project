```python
# take in purchase price, closing costs, and cost for improvements = initial investment
class ROI:
    def __init__(self, initial_investment = {}, income = {}, expenses = {}, cash_pay = 'no'):
        self.initial_investment = initial_investment
        self.income = income
        self.expenses = expenses
        self.cash_pay = cash_pay

# ask for investment numbers:
    def cashInvestment(self):
        
        home_price = input('Expected sale price: $')
        self.initial_investment['Home Price'] = int(home_price)
        cash_pay = input('Will you purchase the home in cash (no mortgage)?') 
        if cash_pay.lower() == 'yes':
            self.cash_pay = 'yes'
            self.initial_investment['Down Payment'] = int(home_price)
        elif cash_pay.lower() == 'no':
            down_pay = input('Expected down payment: $')
            self.initial_investment['Down Payment'] = int(down_pay)
        else:
            print("Please enter 'Yes' or 'No'")
        closing_costs = input('Expected closing costs: $')
        self.initial_investment['Closing Costs'] = int(closing_costs)
        repairs = input('Expected repairs/upgrades cost: $')
        self.initial_investment['Rehab budget'] = int(repairs)
        total_investment = sum(self.initial_investment.values())
        return f'Initial Investment: ${round(total_investment,2)}.'
    
    def monthlyExpenses(self):
        if self.cash_pay == 'no':
            mortgage_payment = input('Expected monthly mortgage payment: $')
            self.expenses['Mortgage Payment'] = int(mortgage_payment)
        prop_tax = input('Expected annual property taxes: $')
        self.expenses['Property Taxes(monthly)'] = int(prop_tax)/12
        prop_insurance = input('Expected annual property insurance premium: $')
        self.expenses['Monthly Property Insurance'] = int(prop_insurance)/12
        hoa = input('Is the property part of an HOA? (Yes/No)')
        if hoa.lower() == 'yes':
            hoa_fees = input('Monthly HOA fee: $')
            self.expenses['HOA Fees'] = int(hoa_fees)
        elif hoa.lower() != 'yes' and hoa.lower() != 'no':
            print("Please enter 'Yes' or 'No'")
        landscaping = input('Will you pay for lawn care? (Yes/No)')
        if landscaping.lower() == 'yes':
           land_fee = input('Expected monthly lawn care cost: $')
           self.expenses['Landscaping cost'] = int(land_fee)
        elif landscaping.lower() != 'yes' and landscaping.lower() != 'no':
            print("Please enter 'Yes' or 'No'")
        building_maint = input('Expected monthly building maintenance costs: $')
        self.expenses['Monthly Building Maintenance'] = int(building_maint)
        vacancy = input('Are you setting aside money in case of a vacancy? 5% is recommended. (Yes/No)')
        if vacancy.lower() == 'yes':
            vacancy_amt = input('Amount being set aside for vacancy: $')
            self.expenses['Vacancy'] = int(vacancy_amt)
        elif vacancy.lower() != 'yes' and vacancy.lower() != 'no':
             print("Please enter 'Yes' or 'No'")
        misc_expenses = input("Enter any additional expenses or '0' if none: $")
        self.expenses['Misc Expenses'] = int(misc_expenses)
        monthly_expenses = sum(self.expenses.values())
        return f'Monthly Expenses are ${round(monthly_expenses,2)}.'
    
    def monthlyIncome(self):
        rental_income = input('How much will you receive in rent each month?')
        self.income['Monthly Rental Income'] = int(rental_income)
        storage = input('Will the tenant pay for storage on the property? (Yes/No)')
        if storage.lower() == 'yes':
            storage_income = input('Expected amount from tenant for storage: $')
            if type(storage_income) == int:
                self.income['Storage Income'] = int(storage_income)
            else:
                print('Please enter a number')
        elif storage.lower() != 'yes' and storage.lower() != 'no':
            print("Please enter 'Yes' or 'No'")
        extra_income = input("Please enter any additional income or '0' if none: $")
        self.income['Additional Income'] = int(extra_income)
        monthly_income = sum(self.income.values())
        return f'The monthly income for the property is ${round(monthly_income,2)}.'
    
    def homeReport(self):
            print(f'================') 
            print(f"Price of Property: ${self.initial_investment['Home Price']}")
            print(f'================')
            for cost, amount in self.initial_investment.items():
                print(f'{cost}: ${round(amount,2)}')
            for expense, amount in self.expenses.items():
                print(f'{expense}: ${round(amount,2)}')
            for income, amount in self.income.items():
                print(f'{income}: ${round(amount,2)}')
            print                       
            
    
    def calculateROI(self):
        #capture Investment info
        self.cashInvestment()
        self.monthlyExpenses()
        self.monthlyIncome()
        self.homeReport()
        monthly_income = sum(self.income.values())
        print(f'Monthly Income = ${round(monthly_income,2)}')
        monthly_expenses = sum(self.expenses.values())
        print(f'Monthly Expenses = ${round(monthly_expenses,2)}')
        roi = ((monthly_income - monthly_expenses)/(sum(self.initial_investment.values())))*100
        return f'Your cash-on-cash ROI is {round(roi,1)}%'
                
house = ROI()
house.calculateROI()                     
                       
                       


#cash flow = income - expenses
# subtract, then multiply by 12 to get yearly cash flow/profit

# divide yearly expense by total investment and multiply x 100
```
